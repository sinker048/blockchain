```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
from collections import defaultdict
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# crawl data from sandbox-api
url = 'https://sandbox-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'5000',
  'convert':'USD'
}
```

```python
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': 'PUT YOUR API KEY',
}
```

```python
session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
# collect the columns names
union_name = set(data['data'][0].keys())
for i in range(1,len(data['data'])):
    temp_name = set(data['data'][i].keys())
    union_name &= temp_name
total_name = list(union_name)
```

```python
# collect the data from data
data_dict = defaultdict(list)
for i in range(len(data['data'])):
    current_data = data['data'][i]
    for name in total_name:
        data_dict[name] += [current_data[name]]
```

```python
df = pd.DataFrame(data_dict)
df
```

```python
# select data from df['quote']
temp_data = defaultdict(list)
total_name = list(df['quote'][i]['USD'].keys())
for i in range(len(data['data'])):
    current_data = df['quote'][i]['USD']
    for name in total_name:
        temp_data['USD_'+name] += [current_data[name]]
    
```

```python
temp_df = pd.DataFrame(temp_data)
temp_df
```

```python
# concat those dataframe and delete df['quote']
df = pd.concat([df, temp_df], axis=1)
del df['quote']
```
